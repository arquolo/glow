```python
import sys

def gf():
    f = sys._getframe()
    # g.gi_suspended, g.gi_running, g.gi_yieldfrom
    try:
        yield
    except GeneratorExit as exc:
        print('genex', exc.args, sys.exc_info(), sys._current_exceptions())
        raise
    except BaseException as exc:
        print('ex', exc.args, sys.exc_info(), sys._current_exceptions())

g = gf()

# no catch, not started yet
# g.throw(GeneratorExit)

# no catch, not started yet
# g.close()

# genex
g.send(None)
g.throw(GeneratorExit)
#
# g.send(None)
# g.close()
```
